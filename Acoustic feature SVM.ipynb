```python
import pandas as pd
import numpy as np
```

```python
join_df = pd.read_csv('Glottal.csv')
join_df = join_df[['filename', 'label']]
```

```python
df = pd.read_csv('SVMDATA.csv')

df.fillna(0, inplace=True)

df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)

# df = pd.merge(df, join_df, on='filename')

df
```

```python
columns = df.columns.tolist()

features = [f for f in filter(lambda x: x not in ['label', 'frameTime','filename'], columns)]
target = ['label']

feature_df = df[features]

X = np.asarray(feature_df)

y = np.asarray(df['label'])

X
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X)

X = scaler.transform(X)

X
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=4)

X_train.shape
```

```python
from sklearn import svm

classifier = svm.SVC(kernel='rbf',C=0.1, gamma=1)

classifier.fit(X_train, y_train)

y_predict = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_predict))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict)
```
